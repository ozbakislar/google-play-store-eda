## 5.8 Price

```python
df[['price']].sample(5)
```

```python
df.price.dtype
```

```python
df.price.unique()
```

```python
df.price.value_counts()
```

```python
df.price = df.price.str.replace('$', '', regex=False)

df.price
```

```python
df.price = pd.to_numeric(df.price, errors="coerce")

df.price
```

```python
df.price.dtype
```

```python
df.price.isnull().sum()
```

```python
df.price.describe().round()
```

```python
df[df.price > 0].price.describe().round()
```

```python
Q1 = df[df.price > 0].price.quantile(0.25)
Q3 = df[df.price > 0].price.quantile(0.75)

IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR

df['high_price'] = df.price > upper

df['high_price']
```

```python
df = df[["app",
         "category",
         "rating",
         "reviews",
         "size_mb",
         "installs",
         "type",
         "price",
         "high_price",
         "content_rating",
         "genres",
         "last_updated",
         "current_ver",
         "android_ver"
       ]]

df
```

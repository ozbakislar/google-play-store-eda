## 5.5 Size

```python
df[['size']].sample(5)
```

```python
df['size'].dtype
```

```python
df['size'].unique()
```

```python
df['size'].nunique()
```

```python
df['size'] = df['size'].str.strip()

df['size']
```

```python
df['size_mb'] = df['size'].transform(lambda x: round(float(x.replace('k', '')) / 1024, 2) if x.endswith('k') else (float(x.replace('M', '')) if x.endswith('M') else None))

df['size_mb']
```

```python
df['size_mb'].dtype
```

```python
df['size_mb'].describe().round()
```

```python
df['size_mb'].isnull().sum()
```

```python
df['size_mb'] = df['size_mb'].fillna(df['size_mb'].median())

df['size_mb']
```

```python
df['size_mb'].isnull().sum()
```

```python
df = df.drop(columns='size')

df
```

```python
df = df[['app',
         'category',
         'rating',
         'reviews',
         'size_mb',
         'installs',
         'type',
         'price',
         'content_rating',
         'genres',
         'last_updated',
         'current_ver',
         'android_ver'
       ]]

df
```

## 5.8 Price

```python
df[['price']].sample(5)
```

```python
df.price.dtype
```

```python
df.price.unique()
```

```python
df.price.value_counts()
```

```python
df.price = df.price.str.replace('$', '', regex=False)

df.price
```

```python
df.price = pd.to_numeric(df.price, errors='coerce')

df.price
```

```python
df.price.dtype
```

```python
df.price.isnull().sum()
```

```python
df.price.describe().round()
```

```python
df[df.price > 0].price.describe().round()
```

```python
Q1 = df[df.price > 0].price.quantile(0.25)
Q3 = df[df.price > 0].price.quantile(0.75)

IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR

df['high_price'] = df.price > upper

df['high_price']
```

```python
df = df[['app',
         'category',
         'rating',
         'reviews',
         'size_mb',
         'installs',
         'type',
         'price',
         'high_price',
         'content_rating',
         'genres',
         'last_updated',
         'current_ver',
         'android_ver'
       ]]

df
```

```python
sns.countplot(data=df, x='year')

plt.title('Anzahl der Anwendungen nach Jahren', fontsize=14, fontweight='bold')
plt.xlabel('Jahr')
plt.ylabel('Anzahl der Anwendungen')
plt.xticks(rotation=45)

plt.show()
```

```python
sns.histplot(data=df, x='rating', hue='type', bins=20, kde=True)

plt.title('Verteilung der Bewertungen (Kostenlos und Kostenpflichtig)', fontsize=14, fontweight='bold')
plt.xlabel('Bewertung')
plt.ylabel('Häufigkeit')

plt.show()
```
